### Data cleaning and preparing for analysis

### Dropping column Unnamed: 0.1

```python
df.drop(columns='Unnamed: 0.1', inplace=True)
```

### Viewing dataframe

```python
df
```

### Renaming column Unnamed: 0 to index

```python
df.rename(columns={'Unnamed: 0': 'index'}, inplace=True)
```

### Viewing DataFrame

```python
df
```

### Specifying the value to replace NaN in the 'Price' column

```python
replacement_value = 'No Info'
```

### Replacing NaN with 'No Info' in the 'Price' column

```python
df['Price'].fillna(replacement_value, inplace=True)
```

### Viewing dataframe

```python
df
```

### Showing unique values in Rocket_Status column

```python
print("Unique Rocket Statuses:", df['Rocket_Status'].unique())
```

### Renaming values in Rocket_Status column

### Defining the mapping for renaming values

```python
status_mapping = {'StatusActive': 'Status: Active', 'StatusRetired': 'Status: Retired'}
```

### Using the map function to replace values in the 'Rocket_Status' column

```python
df['Rocket_Status'] = df['Rocket_Status'].map(status_mapping).fillna(df['Rocket_Status'])
```

### Viewing DataFrame

```python
df
```

### Removing unnamed column

# Identifying and dropping the unnamed index column to avoid confusion

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df.set_index('index', inplace=True)
```

### Viewing dataframe

```python
df
```

### Converting date to datetime format

```python
df['Date'] = pd.to_datetime(df['Date'], utc=True, errors='coerce')
```

### Viewing DataFrame

```python
df
```

### Checking column format types

```python
df.info()
```

### Removing timezone information from date column

```python
df['Date'] = df['Date'].dt.tz_localize(None)
```

### Viewing DataFrame

```python
df
```

### Correcting data types in dataframe

```python
df[['Organisation', 'Location', 'Detail', 'Rocket_Status', 'Mission_Status']] = df[['Organisation', 'Location', 'Detail', 'Rocket_Status', 'Mission_Status']].astype(str)
```

### Checking DataFrame data types

```python
df.info()
```

### Converting Price column to numeric

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

### Checking DataFrame data types

```python
df.info()
```

### Checking if Price column contains , values

```python
contains_comma = df[df['Price'].astype(str).str.contains(',', na=False)]

# Print the length of the resulting DataFrame
print(f"Number of rows with commas in the 'Price' column: {len(contains_comma)}")
```

### Viewing DataFrame

```python
df
```

```python
replacement_value = 'No Info'
```

### Replacing NaN with 'No Info' in the 'Price' column

```python
df['Price'].fillna(replacement_value, inplace=True)
```
